```python

df=pd.read_csv("prices.csv")
df
```

```python
df.plot.scatter(x='area',y='price',title='scatterplot',color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df['price'])
```

```python
reg.predict([[3300]])
```
