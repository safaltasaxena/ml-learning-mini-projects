gradient descent


```python
import numpy as np
#using np since matrix multiplication is convienient with it
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
def gradient_descent(x,y):
  #as we see in 3d graph that we start with some value of m and b initially and
  #then only we take further baby steps.
  m_curr=b_curr=0
  iterations=100 #no of baby steps,u manupulate this when u see constant m and
                   #and const b and mse decreases9
  n=len(x)
  learning_rate=0.08 #manupulate this as mse decreses then move to ite.
  for i in range(iterations):
    y_predicted=m_curr*x+b_curr
    md=(-2/n)*sum(x*(y-y_predicted))
    bd=(-2/n)*sum(y-y_predicted)
    m_curr=m_curr-learning_rate*md
    b_curr=b_curr-learning_rate*bd
    cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
    #cost should reduce after each ite simbolizing that we reduce the error
    print("m {},b {},cost {},iteration {}".format(m_curr,b_curr,cost,i))
gradient_descent(x,y)


```
