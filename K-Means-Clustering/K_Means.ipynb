```python
import pandas as pd
df=pd.read_csv("clustering.csv")
df
```

```python
 from sklearn.cluster import KMeans
 import pandas as pd
 from sklearn.preprocessing import MinMaxScaler
 import matplotlib.pyplot as plt
```

```python
plt.scatter(df[['age']],df['income'])
```

```python
km = KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['age','income']])
y_predicted
#divided data into clusters,assigned diff labels
```

```python
df['cluster']=y_predicted
df
```

```python
f0=df[df.cluster==0]
f1=df[df.cluster==1]
f2=df[df.cluster==2]
plt.scatter(f0['age'],f0['income'],color='green',label='cluster 0')
plt.scatter(f1['age'],f1['income'],color='red',label='cluster 1')
plt.scatter(f2['age'],f2['income'],color='black',label='cluster 2')
plt.xlabel('age')
plt.ylabel('income')
plt.legend()
```

now u see the black ones are in red cluster that happens bcz see the range diff bw y and x axis y is pretty wide x is v small


so we need to do scaling i.e we need to do preprocessing of the data then only we can run our algo

```python
scaler = MinMaxScaler()
scaler.fit(df[['income']])
df['income'] = scaler.transform(df[['income']])
scaler.fit(df[['age']])
df['age'] = scaler.transform(df[['age']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['age','income']])
y_predicted
```

```python
df['cluster']=y_predicted
df
```

```python
km.cluster_centers_
#2 features age and income so 2*3=6
#more of like x and y cord
```

```python
f0=df[df.cluster==0]
f1=df[df.cluster==1]
f2=df[df.cluster==2]
plt.scatter(f0['age'],f0['income'],color='green',label='cluster 0')
plt.scatter(f1['age'],f1['income'],color='red',label='cluster 1')
plt.scatter(f2['age'],f2['income'],color='black',label='cluster 2')
plt.xlabel('age')
plt.ylabel('income')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+',label='centroid')
#all the rows x cord then y cord
plt.legend()
```

here we only had to deal with2 features so we used scatter plot now we will dive in the real life situation where we have 20+ features i.e deciding k from elbow plot

```python
k_range=range(1,10)
sse=[]
for k in k_range:
  km=KMeans(n_clusters=k)
  km.fit(df[['age','income']])
  sse.append(km.inertia_)
  #inertia gives sse directly
```

```python
plt.xlabel('k')
plt.ylabel('sse')
plt.title('elbow plot')
plt.plot(k_range,sse)
#elbow clearly visible at k=3
```
